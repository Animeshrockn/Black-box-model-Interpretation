```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics

from pandas import read_csv

from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
```

```python
# Loading Full data

complete_data = pd.read_csv('data/Used_bikes.csv')
```

```python
complete_data = complete_data.iloc[:10000,:]
complete_data
```

```python
# complete_data.isnull().sum()
# complete_data.value_counts()
```

```python
# Assigning values to variable df_clean

df_clean = complete_data
df_clean
```

```python
# Label Encode categorial values

df_clean['bike_name'] = le.fit_transform(df_clean['bike_name'])
df_clean['city'] = le.fit_transform(df_clean['city'])
df_clean['owner'] = le.fit_transform(df_clean['owner'])
df_clean['brand'] = le.fit_transform(df_clean['brand'])
```

```python
# Finding Correlation between attributes

print(df_clean.shape)

corr_matrix = df_clean.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Features having more than 95% corelation
upper # Rush_att and Rush_yds are the features having correlation more than 95%
```

```python
# Plotting correlation map for better understanding

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df_clean.corr(), vmin=-1, vmax=1, annot=True, cmap='Spectral')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
```

```python
# Checking attributes values to be dropped

to_drop
```

# Global model

```python
# Splitting Data and target 

target   =   df_clean['price'] 
data     =   df_clean.drop('price',axis = 1) 
```

```python
# Splitting the data set into Train and Test data 

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test  =  train_test_split ( data, target, test_size=.1, random_state=40 )
```

```python
# Shape of training and test data

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)

```

```python
# Defining XGBoost Model

from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import numpy as np 

xgb = XGBRegressor(learning_rate=0.01, n_estimators=3000,
                                     min_child_weight=0,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     nthread=-1,
                                     scale_pos_weight=1, seed=27,
                                     random_state=4,loss = 'ls')

```

```python
# Fitting Training data into Model

xgb.fit(x_train.values,y_train.values)
```

```python
# Predicting test values from xgb model

y_pred_xgb = xgb.predict(x_test.values)

```

```python
# Model Score on Training and Testing

# Train
print(xgb.score(x_train.values,y_train.values))

# Test
print(xgb.score(x_test.values,y_test.values))

```

```python
# Feature Importance using Methods defined in Model 

feature_important = xgb.get_booster().get_score(importance_type='weight')
keys = list(feature_important.keys())
values = list(feature_important.values())

# Plotting the Values with their relevance

data1 = pd.DataFrame(data=values, index=keys, columns=["score"]).sort_values(by = "score", ascending=True)
data1.plot(kind='barh')

```

# Clustering

```python
df_clean
```

```python
# Clustering using K-Medoids

from sklearn_extra.cluster import KMedoids

kmed = KMedoids(n_clusters=3, init='k-medoids++',max_iter = 50,random_state=40)    
kmed.fit(df_clean)

labels = kmed.labels_.reshape(10000,1)
result = pd.DataFrame(labels,columns=['label'])
```

```python
# Merging train data with their cluster labels

x_train_cluster = pd.merge(df_clean, result, on=df_clean.index, how='outer')
```

```python
# Dropping key values after joining dataframes

x_train_cluster = x_train_cluster.drop(['key_0'],axis = 1) 
x_train_cluster
```

```python
# Divind data points to respective clusters

cluster1 = x_train_cluster[x_train_cluster['label'] == 0]
cluster2 = x_train_cluster[x_train_cluster['label'] == 1]
cluster3 = x_train_cluster[x_train_cluster['label'] == 2]

```

```python
# Checking cluster shape

print(cluster1.shape)
print(cluster2.shape)
print(cluster3.shape)
```

```python
cluster1
```

# Cluster 1

```python
# Extracting target to new df
t1 = cluster1['price'] 
d1 = cluster1.drop(['price','label'],axis = 1)
```

```python
# Train test split

from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(d1,t1,test_size=.15,random_state=40)
```

```python
print(x_train1.shape)
print(y_train1.shape)
print(x_test1.shape)
print(y_test1.shape)
```

```python
# XGBoost Model

xgb1 = XGBRegressor(learning_rate=0.01, n_estimators=1200,
                                     min_child_weight=0,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     nthread=-1,
                                     scale_pos_weight=1, seed=40,
                                     random_state=4)

xgb1.fit(x_train1.values,y_train1.values)

y_pred_xgb1 = xgb1.predict(x_test1.values)

#here we can check our model score
print(xgb1.score(x_train1.values,y_train1.values))
print(xgb1.score(x_test1.values,y_test1.values))
```

```python
# Feature Importance 

feature_important1 = xgb1.get_booster().get_score(importance_type='weight')
keys = list(feature_important1.keys())
values = list(feature_important1.values())

data_1 = pd.DataFrame(data=values, index=keys, columns=["score"]).sort_values(by = "score", ascending=True)
data_1.plot(kind='barh')
```

```python
# Model - Linear Regression, local model

from sklearn.linear_model import LinearRegression

lr1 = LinearRegression()

lr1.fit(x_train1,y_train1)

y_pred_lr1 = lr1.predict(x_test1)
```

```python
# Score of local model

print(lr1.score(x_train1.values,y_train1.values))
print(lr1.score(x_test1.values,y_test1.values))
```

```python
# Assigning prediction values of XGB and LR models to a1 and b1

a1 = np.array(y_pred_lr1)
b1 = np.array(y_pred_xgb1)
```

```python
# Plotting a1 vs b1 for comparison

dataset1 = pd.DataFrame({'Global': b1, 'Local': list(a1)}, columns=['Global', 'Local'])

dataset1.insert(0, 'New_ID', range(0, 0 + len(dataset1)))

import matplotlib.pyplot as plt

ax = plt.gca()

dataset1.plot(kind='line',x='New_ID',y='Global',ax=ax)
dataset1.plot(kind='line',x='New_ID',y='Local', color='red', ax=ax)
plt.xlim(0,100)
# plt.ylim(0,10000)
plt.ylim(0,100000)
plt.show()
```

# Calculating Global feature importance using our approach

```python
# Calculation part for gloabl coefficinets using cluster coeffs.

gf=[]
for f in range(0,len(lr3.coef_)):
  feature=lr1.coef_[f]+lr2.coef_[f]+lr3.coef_[f]
  feature=feature/4
  gf.append(feature)

gf2=gf
gf2[2]=gf[2]*100
gf2[3]=gf[3]*(-120)

print(gf2)
```

```python
# Plotting feature importance

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,2,2])

plt.xlim(-20000,20000)
plt.ylim(-1,7)
langs = l1
students = gf2
print(len(students))
print(len(langs))

ax.barh(langs,students)
plt.show()
```
